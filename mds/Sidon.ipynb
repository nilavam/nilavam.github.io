```python
#checks whether a given list L of size p is a Sidon set (additively modulo p^2-1)
#time and memory are both O(max(l^2,p^2))
def checkSidon(L,p):
    q = p*p
    l = len(L)
    if(l != p):
        raise Exception("wrong size of list")
    arr = [0]*(q-1)
    for i in range(l):
        for j in range(i+1,l):
            s = (L[i]+L[j])%(q-1)
            arr[s] = arr[s] + 1
    for i in range(q-1):
        if(arr[i]>1):
            print(i,arr[i],'bad')
            return False
    return True
```
