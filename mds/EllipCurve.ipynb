```python
p = 59
q = p*p
F.<t> = GF(p^2)
cutoff = 10
```

```python
desired_order = q-1
i=1
curves = []
gens = []
for a in F:
    for b in F:
        if(i==cutoff+1):
            break
        try:
            E = EllipticCurve(F, [a, b])
            if E.is_smooth():
                n = E.order()
                if (n == desired_order and len(E.gens())==1):
                    curves.append([a,b])
                    gens.append(E.gens())
                    i=i+1
                    break
        except:
            continue
```

```python
for i in range(cutoff):
    print(f"y^2 = x^3 + ({curves[i][0]})x + {curves[i][1]}\n{gens[i][0]}")
    print()
```

```python
num = 0 #which elliptic curve (out of the cutoff=10 elliptic curves)

#get coordinates of generator
x = gens[num][0][0]
y = gens[num][0][1]

#elliptic curve is y^2 = x^3 + ax + b
a = curves[num][0]
b = curves[num][1]

#get the elliptic curve
EC = EllipticCurve(F, [a, b])

#get the generating point on the elliptic curve
B = EC(x,y)

#table to store all points, based on powers of B
table = [EC(0,1,0)]

#generate points on EC by powering B
for i in range(1,q):
    print(i,end=".\t")
    if (i%2==0):
        p = 2*table[i/2]
    else:
        p = 2*table[(i-1)/2] + B
    print(p)
    table.append(p)
```
