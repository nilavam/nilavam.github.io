```python
import numpy as np
def det2(X):
    return(X[0][0]*X[1][1]-X[0][1]*X[1][0])

def det3(X):
    X1 = np.array([[X[1][1], X[1][2]],[X[2][1],X[2][2]]])
    X2 = np.array([[X[1][0], X[1][2]],[X[2][0],X[2][2]]])
    X3 = np.array([[X[1][0], X[1][1]],[X[2][0],X[2][0]]])
    return(X[0][0]*det2(X1) - X[0][1]*det2(X2) + X[0][2]*det2(X3))

def check_concurrent(r,X,ls):
    if(0==det3(np.array([X[r[0]][r[1]],X[r[2]][r[3]],X[r[4]][r[5]]]))):
        #print("\t\t",end="\t\t\t\t\t")
        return([ls[i] for i in r])
    #print()
```
